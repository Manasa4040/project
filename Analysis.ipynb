```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd
EV_final_path = Path("Final_Resource/EV_final.csv")
EV_final_df = pd.read_csv(EV_final_path)
EV_final_df.head()
```

```python
#Top 5 States with most ev registered
EV_final_df.sort_values("Total EV", ascending= False).head()
```

```python
#top 5 states with highest percentage growth  in sales between year 2020-2021)
new_values=EV_final_df["YOY(2020-2021)Growth Percentage"].str.strip("%").astype(float)
EV_final_df["YOY(2020-2021)Growth Percentage"]=new_values
EV_final_df.sort_values(by = "YOY(2020-2021)Growth Percentage",ascending = False).head()
```

```python
#top 5 states with the highest number of charging outlets
EV_final_df.sort_values(by = "Total charging outlets",ascending = False).head()
```

```python
#top 5 states with the best Electric vehicles to charger ports ratio
EV_final_df.sort_values(by = "EV:Charger outlets",ascending = False).head()
```
