```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd
import requests
import json
from pathlib import Path
from pprint import pprint

# created path for the csv files
sales_file = Path("Resources/Ev_Sales_US_states.csv")
charging_file = Path("Resources/chargingpoints_state_us.csv")
fuel_file = Path("Resources/Average_fuel_price.csv")
charging_outlets = Path("Resources/df_2019_2021_clean.csv")
state_capital_file = Path("Resources/usa-state-capitals.csv")

#Load the CSV file created in Part 1 into a Pandas DataFrame
EV_data = pd.read_csv(sales_file)
charging_data = pd.read_csv(charging_file)
fuel_data = pd.read_csv(fuel_file)
charging_outlets_data = pd.read_csv(charging_outlets)
State_capital_data = pd.read_csv(state_capital_file)


# merged data files
EV_Charging_df = pd.merge(EV_data,charging_data,how = "left", on = "State")
EV_Charge_fuel_df= pd.merge(EV_Charging_df,fuel_data,how="left",on="State")
EV_charge_fuel_charging_outlets_df=pd.merge(EV_Charge_fuel_df,charging_outlets_data,how="left",on="State")
EV_charge_fuel_charging_outlets_capital_df=pd.merge(EV_charge_fuel_charging_outlets_df,
                                                    State_capital_data,how="left",on="State")

EV_charge_fuel_charging_outlets_capital_df
```

```python
#dropped Nan Value Rows
EV_charge_fuel_charging_outlets_capital_clean_df= EV_charge_fuel_charging_outlets_capital_df.dropna()
EV_charge_fuel_charging_outlets_capital_clean_df
```

```python
#dropped Columns Not needed for analysis
EV_clean_df= EV_charge_fuel_charging_outlets_capital_clean_df.drop(["Level 1 Ports (1)","Level 2 Ports (1)",
                                                                   "DCFC Ports (1)","Percent of total EVs (2021)",
                                                                   "Percent of total EVs (2020)"," Mid-Grade  ",
                                                                   " Premium  "," Diesel  ","Index","Longitude","Latitude"],axis = 1)
EV_clean_df.head()
```

```python
#renamed columns for easier readability
EV_final_df = EV_clean_df.rename(columns = {"# of EV registrations (2019)":"Registrations of EV 2019",
                                            "# of EV registrations (2020)":"Registrations of EV 2020",
                                            "# of EV registrations (2021)":"Registrations of EV 2021",
                                            "Total EVs (2)":"Total EV","YOY Growth":"YOY(2020-2021)Growth Percentage", 
                                            "Total Ports (1)":"Total charging outlets","Ratio: EVs to Charger Ports":"EV:Charger outlets",
                                            " Regular  ":"Average fuel price(current)"})
EV_final_df.head()
```

```python
#recalculating the total EV to reflect the sum of registrations made in (2019,2020,2021)
#recalculating the total Charging outlets to reflect the sum of charging outlets in (2019,2020,2021)
#coverting the data types of the columns to float for calculation bases
EV_final_df[["Registrations of EV 2019","Registrations of EV 2020","Registrations of EV 2021"]].astype(float)
EV_final_df["2019 - No. EV charge outlets"]=EV_final_df["2019 - No. EV charge outlets"].str.replace(",","").astype(float)
EV_final_df["2020 - No. EV charge outlets"]=EV_final_df["2020 - No. EV charge outlets"].str.replace(",","").astype(float)
EV_final_df["2021 - No. EV charge outlets"]=EV_final_df["2021 - No. EV charge outlets"].str.replace(",","").astype(float)
#forming empty list for each of the columns we want to modify
Total_EV = []
Total_Charging_outlets = []
EV_ratio_Charging_outlets = []
#calcualting the calculation on each row using a for loop and appending the emptylists
for index, row in EV_final_df.iterrows():
    total_ev=(row["Registrations of EV 2019"])+(row["Registrations of EV 2020"])+(row["Registrations of EV 2021"])
    total_outlets = (row["2019 - No. EV charge outlets"]+row["2020 - No. EV charge outlets"]+row["2021 - No. EV charge outlets"])
    ev_ratio_charging_oulets = round(total_ev/total_outlets,2)
    Total_EV.append(total_ev)
    Total_Charging_outlets.append(total_outlets)
    EV_ratio_Charging_outlets.append(ev_ratio_charging_oulets)
#modifying the columns with new value
EV_final_df["Total EV"]=Total_EV
EV_final_df["Total charging outlets"] = Total_Charging_outlets
EV_final_df["EV:Charger outlets"] = EV_ratio_Charging_outlets
EV_final_df.head()
```

```python
#rearranging the columns
EV_final_df= EV_final_df[["State","Registrations of EV 2019","Registrations of EV 2020","Registrations of EV 2021","Total EV",
                          "YOY(2020-2021)Growth Percentage","2019 - No. EV charge outlets","2020 - No. EV charge outlets",
                          "2021 - No. EV charge outlets","Total charging outlets","EV:Charger outlets",
                          "Average fuel price(current)","Capital","ST"
                         ]]
EV_final_df.head()
```

```python
#creating columns for latitude and longitude geoplots

geoapify_key = "4de3692f5d3041f8aeaf76cc55cf1d50"
# Define the API parameters
params = {
    "apiKey":geoapify_key,
    "format":"json",
}

# Set the base URL
base_url = "https://api.geoapify.com/v1/geocode/search"

#creating empty columns for latitude and longitude
EV_final_df["Lat"]=""
EV_final_df["Lon"]=""
EV_final_df.head()
```

```python
print("Starting state search")

# Loop through the EV_final DataFrame and search coordinates for each capital_city,state,country
for index, row in EV_final_df.iterrows():

   
    State =row["Capital"]+","+row["ST"]+",United States of America"

    # Add the state to the parameters
    params["text"] = f"{State}"
    # Make the API request
    response = requests.get(base_url, params=params)
    
    # Convert reponse to JSON
    response_data = response.json()

    # Extract latitude and longitude
    EV_final_df.loc[index, "Lat"] = response_data["results"][0]["lat"]
    EV_final_df.loc[index, "Lon"] = response_data["results"][0]["lon"]
    
    # Log the search results
    print(f"Coordinates for {State} fetched...")

# Display sample data to confirm that the coordinates appear
EV_final_df.head()
```
