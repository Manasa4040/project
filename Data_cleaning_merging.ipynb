```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# created path for the csv files
sales_file = Path("Resources/Ev_Sales_US_states.csv")
charging_file = Path("Resources/chargingpoints_state_us.csv")
fuel_file = Path("Resources/Average_fuel_price.csv")
lat_lng_file = Path("Resources/statelatlong.csv")
charging_outlets = Path("Resources/df_2019_2021_clean.csv")

#Load the CSV file created in Part 1 into a Pandas DataFrame
EV_data = pd.read_csv(sales_file)
charging_data = pd.read_csv(charging_file)
fuel_data = pd.read_csv(fuel_file)
lat_lng_data = pd.read_csv(lat_lng_file)
charging_outlets = pd.read_csv(charging_outlets)

# merged data files
EV_Charging_df = pd.merge(EV_data,charging_data,how = "left", on = "State")
EV_Charge_fuel_df= pd.merge(EV_Charging_df,fuel_data,how="left",on="State")
EV_charge_fuel_cord_df=pd.merge(EV_Charge_fuel_df,lat_lng_data,how="left",on="State")
EV_charge_fuel_cord_chargingoutlets_df =pd.merge(EV_charge_fuel_cord_df,charging_outlets,how="left",on="State")

EV_charge_fuel_cord_chargingoutlets_df
# merging the average fuel_data
```

```python
#dropped Nan Value Row
EV_charge_fuel_cord_chargingoutlets_clean_df= EV_charge_fuel_cord_chargingoutlets_df.dropna()
EV_charge_fuel_cord_chargingoutlets_clean_df
```

```python
#dropped Columns Not needed for analysis
EV_clean_df= EV_charge_fuel_cord_chargingoutlets_clean_df.drop(["Level 1 Ports (1)","Level 2 Ports (1)","DCFC Ports (1)","Percent of total EVs (2021)",
                "Percent of total EVs (2020)"," Mid-Grade  "," Premium  "," Diesel  ","Index"],axis = 1)
EV_clean_df.head()
```

```python
#renamed columns for easier readability
EV_final_df = EV_clean_df.rename(columns = {"# of EV registrations (2019)":"Sales of EV 2019","# of EV registrations (2020)":"Sales of EV 2020",
                     "# of EV registrations (2021)":"Sales of EV 2021","Total EVs (2)":"Total EV(current)",
                      "YOY Growth":"YOY(2020-2021)Growth Percentage", "Total Ports (1)":"Total charging ports(current)",
                      "Ratio: EVs to Charger Ports":"EV:Charger ports(current)"," Regular  ":"Average fuel price(current)"})
EV_final_df.head()
```

```python
EV_final_df.to_csv("Final_Resource/EV_final.csv",index_label = "Index")
```
