```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd
import requests
import json
from pathlib import Path
from pprint import pprint

# Dependencies and Setup
EV_final_path = Path("Final_Resource/EV_final.csv")
EV_final_df = pd.read_csv(EV_final_path)
EV_final_df.head()
```

```python
#Top 5 States with most ev registered
EV_final_df.sort_values("Total EV", ascending= False).head()
```

```python
#top 5 states with highest percentage growth  in sales between year 2020-2021)
new_values=EV_final_df["YOY(2020-2021)Growth Percentage"].str.strip("%").astype(float)
EV_final_df["YOY(2020-2021)Growth Percentage"]=new_values
EV_final_df.sort_values(by = "YOY(2020-2021)Growth Percentage",ascending = False).head()
```

```python
#top 5 states with the highest number of charging outlets
EV_final_df.sort_values(by = "Total charging outlets",ascending = False).head()
```

```python
#top 5 states with the best Electric vehicles to charger ports ratio
EV_final_df.sort_values(by = "Total EV",ascending = False).head()
```

```python
# Sort the DataFrame by 'Total EV' in descending order
sorted_df = EV_final_df.sort_values(by='Total EV', ascending=False)

# Select the top five states by 'Total EV'
top_states_df = sorted_df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8)) 

states = top_states_df['State']
years = [2019, 2020, 2021]
bar_width = 0.2
index = range(len(states))

for year in years:
    ev_registrations = top_states_df[f'Registrations of EV {year}']
    plt.bar([pos + bar_width * years.index(year) for pos in index], ev_registrations, width=bar_width, label=year)

plt.xlabel('States')
plt.ylabel('EV Registrations')
plt.title('Year-wise Trend Analysis for Top Five States')
plt.xticks([pos + bar_width for pos in index], states)
plt.legend()
plt.grid(True)
plt.savefig('Output_Data/Year-wise Trend Analysis for Top Five States.png')
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 8))

states = top_states_df['State']
years = [2019, 2020, 2021]
bar_width = 0.2
index = range(len(states))

for i, year in enumerate(years):
    ev_registrations = top_states_df[str(year) + ' - No. EV charge outlets']
    plt.bar([pos + bar_width*i for pos in index], ev_registrations, width=bar_width, label=str(year))

plt.xlabel('States')
plt.ylabel('Charging Outlets')
plt.title('Trend of Charging Outlets for Top Five States')
plt.xticks([pos + bar_width for pos in index], states)
plt.legend()
plt.grid(True)
plt.savefig('Output_Data/bar graph Trend of Charging Outlets for Top Five States.png')
plt.show()
```

```python
# Plotting the state-wise trend analysis
plt.figure(figsize=(12, 8))

years = [2019, 2020, 2021]

for state in top_states_df['State']:
    state_data = EV_final_df[EV_final_df['State'] == state]
    ev_registrations = state_data[['Registrations of EV 2019', 'Registrations of EV 2020', 'Registrations of EV 2021']].values[0]
    plt.plot(years, ev_registrations, marker='o', label=state)

plt.xlabel('Year')
plt.ylabel('EV Registrations')
plt.title('State-wise Trend Analysis for Top Five States')
plt.xticks(years)
plt.legend()
plt.grid(True)
plt.savefig('Output_Data/line graph state_wise_trend_analysis.png')
plt.show()

```
