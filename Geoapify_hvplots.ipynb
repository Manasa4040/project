```python
%%capture --no-display

# Configure the map plot
map_plot_1 = EV_final_df.hvplot.points(
    "Lon",
    "Lat",
    geo = True,
    tiles = "OSM",
    frame_width = 700,
    frame_height = 500,
    size= "Total EV",
    color = "State",
    scale = 0.05,
    title = "Total_ev_registrations(2019-2021)",
    hover_cols=["Registrations of EV 2019","Registrations of EV 2020","Registrations of EV 2021"]
    
)
hvplot.save(map_plot_1,'Output_Data/EV_registrations.png')
map_plot_1
```
