```python
## Dependencies and Setup
import random
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
import requests
import json
from pathlib import Path
from pprint import pprint
import scipy.stats as st

# created path for the csv files
EV_final_path = Path("Final_Resource/EV_final.csv")
# Read the mouse data and the study results
final_file_df = pd.read_csv(EV_final_path)
#display data
final_file_df
```

```python
#top 5 states with the highest number of charging outlets
final_file_df.sort_values(by = "Total charging outlets",ascending = False).head()
```

```python
#top 5 states with the best Electric vehicles to charger ports ratio
final_file_df.sort_values(by = "EV:Charger outlets",ascending = False).head()
```

```python
#Top 5 States with most ev registered
final_file_df.sort_values("Total EV", ascending= False).head()
```

```python
#bar graphs for the trend of ev sales over the years.
Reg_df = final_file_df[['Registrations of EV 2019', 'Registrations of EV 2020', 'Registrations of EV 2021']].mean()

Reg_df.plot(kind= 'bar', figsize=(4,4))
plt.xlabel('Registration year')
plt.ylabel('Number of Ev Registrations')
plt.title('Triend of EV Registrations over the years')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
top5_EV_df=final_file_df.sort_values(by = "YOY(2020-2021)Growth Percentage",ascending = False).head()
top5_EV_df
```

```python
#top 5 states with highest percentage growth  in sales between year 2020-2021)
new_values=final_file_df["YOY(2020-2021)Growth Percentage"].str.strip("%").astype(float)
final_file_df["YOY(2020-2021)Growth Percentage"]=new_values
final_file_df.sort_values(by = "YOY(2020-2021)Growth Percentage",ascending = False).head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure( figsize=(10,8))
#plotdata.plot(kind="bar",figsize=(15, 8))


states=top5_EV_df['State']
year= [2019, 2020,2021]
bar_width=0.2
index= range(len(states))
for year in years:
    ev_registrations=top5_EV_df[f'Registrations of EV {year}']
    plt.bar([pos+ bar_width* years.index(year)for pos in index], ev_registrations,width= bar_width, label=year)
    
   # plotdata = pd.DataFrame(index=range(len(states))
    
    plt.xlabel('States')
    
    plt.ylabel('year-wise trend analysis for top five states')
    
    plt.title('Trend of EV Registrations over the years')
    plt.xticks([pos + bar_width for pos in index],states)
    plt.grid(True)
    plt.savefig('trend of Ev registrations over the years_plot.png')
    plt.show()
    
    
```

```python
#bar graph for the trend of charging outlets over the years. 
    
Charging_df = final_file_df[['2019 - No. EV charge outlets', '2020 - No. EV charge outlets', '2021 - No. EV charge outlets']].mean()

Charging_df.plot(kind= 'bar', figsize=(4,4))
plt.xlabel('charging outlets per year')
plt.ylabel('Number of charging outlets')
plt.title('Trend of chargingoutlets over the years')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()    
    
```
